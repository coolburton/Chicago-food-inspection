```python
engine = create_engine("postgres+psycopg2://postgres:PASSWORD@localhost:5432/violation_codes")
conn = engine.connect()
```

```python
sql_violationData = pd.read_sql("SELECT * FROM violation_codes", conn)
```

```python
sql_violationData.dtypes
```

```python
violationarrayformatch = []
codearrayformatch = []
for violation in restaurants_pd['Violations']:
    for code in sql_violationData['violation_code']:
        if code in violation:
            violationarrayformatch.append(violation)
            codearrayformatch.append(code)
```

```python
violationmatchingcolumns = ["violation_code", "Violations"]
violation_matching = pd.DataFrame(columns=(violationmatchingcolumns))
violation_matching["violation_code"] = codearrayformatch
violation_matching["Violations"] = violationarrayformatch
```

```python
violation_combined = violation_matching.merge(sql_violationData, on="violation_code")
```

```python
addedInfo_restaurants_pd = restaurants_pd.merge(violation_combined, on="Violations", how="left")
```

```python
addedInfo_restaurants_pd = addedInfo_restaurants_pd.drop_duplicates(subset=["Violations", "Inspection_ID"])
```
